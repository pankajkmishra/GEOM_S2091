```python
import numpy as np

# Define the states and transition matrix
states = ["Sunny", "Cloudy", "Rainy"]
transition_matrix = np.array([
    [0.7, 0.2, 0.1],  # P(Sunny -> *)
    [0.3, 0.4, 0.3],  # P(Cloudy -> *)
    [0.2, 0.3, 0.5]   # P(Rainy -> *)
])

# Function to simulate the Markov chain
def simulate_markov_chain(start_state, steps):
    state_index = states.index(start_state)
    states_visited = [start_state]
    
    for _ in range(steps):
        state_index = np.random.choice(range(3), p=transition_matrix[state_index])
        states_visited.append(states[state_index])
    
    return states_visited

# Simulate the Markov chain starting from 'Sunny'
simulation = simulate_markov_chain("Sunny", 1000)
simulation
```

```python
import matplotlib.pyplot as plt

# Running an extended simulation to observe long-term behavior
long_simulation = simulate_markov_chain("Cloudy", 10000)

# Count the frequency of each state
state_counts = {state: long_simulation.count(state) for state in states}

# Plotting the histogram
plt.figure(figsize=(8, 5))
plt.bar(state_counts.keys(), state_counts.values(), color=['gold', 'gray', 'blue'])
plt.title('Frequency of Each State in a Long-term Simulation')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.show()

state_counts
```
