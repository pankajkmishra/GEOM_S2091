```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function with m1 and m2
def fun(m1, m2):
    return (3 * (1 - m1)**2 * np.exp(-(m1**2) - (m2 + 1)**2) -
            10 * (m1 / 5 - m1**3 - m2**5) * np.exp(-m1**2 - m2**2) -
            1/3 * np.exp(-(m1 + 1)**2 - m2**2))

# Create a grid for m1 and m2
m1 = np.linspace(-3, 3,100)
m2 = np.linspace(-3, 3,100)
M1, M2 = np.meshgrid(m1, m2)

# Evaluate the function at each point on the grid
Z = fun(M1, M2)

# Find the minimum using a grid search
minValue = np.min(Z)
minIndex = np.argmin(Z)
minM1Index, minM2Index = np.unravel_index(minIndex, Z.shape)
minM1 = M1[minM1Index, minM2Index]
minM2 = M2[minM1Index, minM2Index]

# Plotting the function
plt.figure(figsize=(10, 8))
contour = plt.contourf(M1, M2, Z, levels=50, cmap='viridis')
plt.colorbar(contour)
plt.scatter(minM1, minM2, color='red', marker='o', label=f'Minimum ({minM1:.2f}, {minM2:.2f})')
plt.legend()
plt.title('Error function')
plt.xlabel('m1')
plt.ylabel('m2')
plt.show()

# Print the minimum location and value
print(f'The minimum is at m1 = {minM1:.4f}, m2 = {minM2:.4f} with a value of {minValue:.4f}')
```
