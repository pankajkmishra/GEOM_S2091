```python
import numpy as np
import matplotlib.pyplot as plt

# Generating samples from a normal distribution
mu, sigma = 175, 20  # mean and standard deviation
normal_samples = np.random.normal(mu, sigma, 1000)

# Generating samples from a uniform distribution
low, high = 0, 300  # range of values
uniform_samples = np.random.uniform(low, high, 1000)

# Plotting the histograms
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Normal Distribution
axs[0].hist(normal_samples, bins=30, color='blue', alpha=0.7)
axs[0].set_title('Normal Distribution\nMean = 175, Std Dev = 20')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')

# Uniform Distribution
axs[1].hist(uniform_samples, bins=30, color='green', alpha=0.7)
axs[1].set_title('Uniform Distribution\nRange = [0, 300]')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
